***Train test split***

The dataset has to split into training data and testing data.Doing it manually is not efficient. We have a train_test_split method from sklearn which simplifies the job for us.

It can split the data based on the ratios we input using test_size parameter.

Stratify parameter ensures the dataset is split in a balanced manner(equal positives and negatives) in both testing and training sets.

Random_state parameter ensures the same set of data is considered for every execution of the model so as to provide reproducible results.

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('class',axis=1)
Y = data['class']
```

*Exploring test_size parameter:*

```python
#magical line of code to split 
X_train, X_test ,Y_train, Y_test = train_test_split(X,Y,test_size=0.1)
```

```python
#Splitting data in 80:20 ratio - test_size=0.2 
X_train, X_test ,Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
print(X.shape,X_train.shape,X_test.shape)
print(Y.shape,Y_train.shape,Y_test.shape)
print(Y.mean(),Y_train.mean(),Y_test.mean())
```

*Exploring stratify parameter:*

Stratify parameter ensures the dataset is split in a balanced manner(equal positives and negatives) in both testing and training sets.

Note the difference in mean values of Y_train,Y_test before and after using stratify.When stratify=Y the mean of train and test tends to be almost same

```python
#To ensure balnced split of data use stratify=Y
X_train, X_test ,Y_train, Y_test = train_test_split(X,Y,test_size=0.2, stratify = Y)
print(X.shape,X_train.shape,X_test.shape)
print(Y.shape,Y_train.shape,Y_test.shape)
print(Y.mean(),Y_train.mean(),Y_test.mean())
```

*Exploring random_state parameter:*

```python
#Random state parameter ensures reproducible results by controlling the randomness in which data is split in every iteration
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=0)
print(X_train.mean(),X_test.mean(),X.mean())
```

***Exploring shuffle parameter:***

It tells whether or not to shuffle the data before splitting. If shuffle=False then stratify must be None.
shuffle=True by default

```python
X_train, X_test ,Y_train, Y_test = train_test_split(X,Y,test_size=0.2, stratify = None,shuffle=True)
print(X_train.mean(), X_test.mean, X.mean())
```

***Binarization***

The domain of features can be huge.The values can be transformed into a scale of 0 and 1 for ease and to avoid a complex model if the purpose of the model can still be met.

```python
#binarization of input values
X_binarised_train = X_train.apply(pd.cut , bins=2,labels=[1,0])
X_binarised_test = X_test.apply(pd.cut,bins=2, labels=[1,0])
```

```python
#Converting dataframes into numpy array
X_binarised_train = X_binarised_train.values
X_binarised_test = X_binarised_test.values
```

***Training the model***

```python
#Simple training model 
 
#Training the model
b_optimal=0
max_accuracy=0
for b in range (X_binarised_train.shape[1]+1):
    accurate_rows =0

    for x,y in zip(X_binarised_train, Y_train):
        #Predicting the target value
        y_pred=(np.sum(x)>=b)
        #Checking if prediction is correct and counting the number of accurate predictions
        accurate_rows +=(y == y_pred )
    #Calculating the accuracy for every value of b
    # Accuracy = Num of rows predicted correctly/total number of rows
    accuracy=(accurate_rows/X_binarised_train.shape[0])*100

    print("b="+str(b),end=" ")
    print("No of Accurate rows= "+str(accurate_rows),end=" ")
    print("Accuracy= "+str(accuracy)+str("%\n"))
      
    #Finding the optimal value of b
    if(accuracy>max_accuracy):
      max_accuracy=accuracy
      b_optimal=b

print("***********************************************")
print("Maximum Training Accuracy= "+str(max_accuracy)+str("*\n"))
print("Optimal b value while trainng= "+str(b_optimal)+str("\n"))
print("***********************************************")
```

***Testing the model***

```python
#Testing the model
from random import randint
max_accuracy=0
i=randint(0,X_binarised_test.shape[0])
for b in range (X_binarised_test.shape[1]+1):
    accurate_rows =0

    for x,y in zip(X_binarised_test, Y_test):
        y_pred=(np.sum(x)>=b)
        accurate_rows +=(y == y_pred )
    
    #Calculating the accuracy for every value of b
    # Accuracy = Num of rows predicted correctly/total number of rows
    accuracy=(accurate_rows/X_binarised_test.shape[0])*100

    print("b="+str(b),end=" ")
    print("No of Accurate rows= "+str(accurate_rows),end=" ")
    print("Accuracy= "+str(accuracy)+str("%\n"))
      
    #Finding the optimal value of b
    if(accuracy>max_accuracy):
      max_accuracy=accuracy
      b_optimal=b

print("***********************************************")
print("Maximum Testing Accuracy= "+str(max_accuracy)+str("%\n"))
print("Optimal b value while testing= "+str(b_optimal)+str("\n"))
print("***********************************************")

    

```
